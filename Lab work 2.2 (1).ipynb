* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import numpy as np
from collections import Counter
import statistics
# генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

sum_list = sum(list_with_random_numbers)/len(list_with_random_numbers)
list_with_random_numbers.sort()
if len(list_with_random_numbers) % 2 == 0:
    a = len(list_with_random_numbers) // 2
    result = (list_with_random_numbers[a-1] + list_with_random_numbers[a])/2
    print("Mедіанa послідовності 2: ",result)

count_dict = dict(Counter(str(list_with_random_numbers)))

frequency_dict = {}
for key in count_dict.keys():
    frequency_dict[key] = count_dict[key] / len(list_with_random_numbers)

mean = statistics.mean(list_with_random_numbers)

stdev = statistics.stdev(list_with_random_numbers)


print("Mедіанa послідовності 1: ",np.median(list_with_random_numbers))
print("Мінімальне значення: ",min(list_with_random_numbers))
print("Максимальне значення: ",max(list_with_random_numbers))
print("Cереднє значення послідовності: ",sum_list)
print("Частотa кожної цифри в послідовності: ",frequency_dict)
#print("Mean:", mean)
print("Середнє квадратичне відхилення", round(stdev,3))


                    
```
